##01-1 인공지능을 소개합니다

**인공지능(artificial Interlligence)** : 사람의 지능을 만들기 위한 시스템이나 프로그램<br>ex) 알파고(바둑), 음성비서(사람의 음성질문에 답할 수 있는 프로그램)

인공기술의 발전으로 실제 사람이 말하듯 가짜 영상도 제작가능 > '사람을 완벽하게 대체할 수도 있는 기술'이란 오해

###**인공지능의 종류**


####1.   강인공지능


> '사람과 구분이 안 될 정도로 강한 성능을 가진 인공지능' <br>ex)영화속에 등장하는 인공지능 -자비스(아이언맨) <br>현재는 만들 수 있는 방법이 없음

####2.   약 인공지능


> 현재 우리가 사용하는 인공지능 기술<br> 특정 영역에서 작업을 수행하는 인공지능<br>ex)운전보조, 질문답변, 검색
<br>머신러닝, 딥러닝은 이 곳에 포함


###**머신러닝과 딥러닝 그리고 인공지능의 관계**
<br> 머신러닝과 딥러닝은 약 인공지능에 포함되는 기술, 딥러닝은 머신러닝에 포함되는 기술
<br>머신러닝은 알고리즘의 종류에 따라 세부적으로 나누어짐
<br> ex) 딥러닝 : 인공신경망 알고리즘을 사용하여 만든 것

##01-2 머신러닝을 소개합니다

머신러닝 : 기계학습
<br>
###**머신러닝은 스스로 규칙을 수정**

'학습' : 머신러닝과 딥러닝에서 데이터 규칙을 컴퓨터 스스로 찾아내는 것
<br> 전통적 프로그램 : 사람이 정한 규칙대로 동작 사람이 규칙을 정하여 프로그래밍, 사람이 프로그램의 실행 결과를 보며 수정
<br> 머신러닝 : 사람이 만든 프로그램 이지만 규칙을 스스로 찾아 수정(훈련)
<br>
###**머신러닝의 학습 방식을 이해합니다**

머신러닝의 학습 방식에 따른 분류


*   지도학습(supervised learning)
*   비지도학습(unsupervised learning)
*   강화학습(reinforcement learning)

이 책에서 다루는 예제들은 모두 지도학습
<br><br>
####**지도 학습은 입력과 타깃으로 모델을 훈련시킵니다**
모델 : 학습을 통해 만들어진 프로그램
<br>훈련데이터 : 모델을 훈련시키기 위해 사용하는 데이터, 입력과 타깃으로 구성 됨
<br>입력 : 모델이 풀어야 할 문제
<br>타깃 : 모델이 맞춰야 할 정답
<br>문제에 대한 답을 주는 방법으로 모델을 훈련, 모델은 새로운 입력에 대한 예측을 만들어냄 <br>즉 '학습을 통해 예측하는 프로그램'
<br>지도 학습은 기존의 데이터를 통해 모델을 학습시키고 학습시킨 모델로 새로운 입력에 대한 예측  날씨 예측, 스팸 이메일 분류
<br>지도학습에서 가장 어려운 일은 모델을 훈련시키는 것이 아니라 훈련 데이터를 만드는 것
<br>데이터의 정확도에 따라 모델의 정확성이 떨어짐
<br><br>
####**비지도 학습은 타깃이 없는 데이터를 사용합니다**
<br>비지도학습 : 타깃이 없는 훈련 데이터를 사용(그룹을 지정하는 상황)
<br>ex)군집(clustering)
<br>훈련 데이터에 타깃이 없음으로 모델의 훈련 결과를 평가하기가 어려움
<br><br>
####**강화 학습은 주어진 환경으로부터 피드백을 받아 훈련합니다**
<br>강화학습은 머신러닝 알고리즘으로 에이전트라는 것을 훈련
<br>훈련된 에이전트는 특정 환경에서 최적화된 행동을 수행하고 수행에 대한 '보상'과 '현재 상태'를 받음 '최대한 많은 보상을 받는 것'을 목표로함 <br>아주 많은 행동을 수행하며 학습을 이룸
<br>ex)딥마인드의 알파고, 온라인 광고
<br> 강화학습의 대표적인 알고리즘 : Q러닝, SARSA, 인공신경망을 사용한DQN
<br><br>
###**규칙이란 가중치와 절편을 말합니다**

대부분의 머신러닝 알고리즘은 훈련 데이터와 규칙의 관계를 식으로 표현 할 수 있음
<br>ex)'오늘 습도가 0.6이상이면 내일 비가온다'를 규칙으로 정함

\\( 1.5*x+ 0.1 =y \\) 로 식을 만들 수 있음 y가 1이상이면 비가 온다고 예측
<br>x:입력 1.5 : 가중치(곱하는 수), 0.1 : 절편(더하는 수) 
<br>딥러닝은 10만~100만개의 가중치와 절편을 가짐 
<br><br>
###**모델은 머신러닝의 수학적 표현입니다**

앞에서 만든 수학식이 모델
<br>모델 파라미터 : 가중치와 절편
<br>이 책의 실습에서는 클래스를 이용하여 구현함으로 이를 이용한 객체를 모델로 생각
<br><br>
###**손실 함수로 모델의 규칙을 수정합니다**

손실 함수 : 모델의 규칙을 수정하는 기준이 되는 함수
<br>ex)모델이 예측한 값과 타깃의 값의 차이를 계산하는 함수


**최적화 알고리즘으로 손실 함수의 최솟값을 찾습니다**
<br>손실 함수의 최솟값을 효율적으로 찾는 방법 필요 
<br>최적화 알고리즘 : 최솟값을 효율적으로 찾는 방법
