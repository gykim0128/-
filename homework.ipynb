# 과제

10진수의 소수인 0.1을 2진수의 숫자로 바꾸기 위해서 일단 10진수의 각 자리가 10^n을 나타내듯 2진수의 각 자리도 2^n을 나타낸다 즉 소수점 아래의 자리수는 2^(-n)자리를 나타내게 된다. 소수점 첫째 자리부터 0.5 0.25 0.125 *** 이런 식으로 나타내어진다. 

인터넷에서 찾아본니 예를들어 0.375가 있다면
곱하기 2를 하면 0.750되고 소숫점위로올라가는건 없으니 0
0.750곱하기 2 = 1.500 여기서 소숫점위로가는건 1
방금결과에서 1을 제외한 0.500곱하기 2 결과는  1
결국 0.375는 2진수로 0.011로 나타내어진다.

[출처] 소수/음수 10진수 2진수 변환|작성자 아울리에

구해야하는 0.1을 이러한 식으로 나타내보면 2를 곱하면 0.2 다시 2를 곱하면 0.4 0.8 1.6 순으로 되고 이때 첫번째 숫자가 처음으로 0.0001이 등장한다. 다시 0.6에서 2를 곱하면 1.2가 되고 두번째 자리수가 나타내어 진다. 이 다음은 처음에 2를 곱한것과 같기 때문에 순환소수로 나타내어지는데 0.0001100110011*** 으로 무한 소수가 됩니다
